```python
import pandas as pd 
import numpy as np
import re
import string
import nltk
from nltk.corpus import stopwords  
stop = stopwords.words('english')
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer 
ps = PorterStemmer() 
from gensim.models import Word2Vec
modelSize = 300
from sklearn.manifold import TSNE
from sklearn.preprocessing import scale
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.calibration import CalibratedClassifierCV
vectorizer = TfidfVectorizer()
cv= CountVectorizer(binary=True, max_df=0.95)
cb= CountVectorizer(binary=False, max_df=0.95)
import matplotlib
import matplotlib.pyplot as plt
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 15),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)

#
### load files
#
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
dev = pd.read_csv('dev.csv')

# 
### get text and sentiment from files
#
def getText(dataframe):
    return dataframe['text']

def getSentiment(dataframe):
    return dataframe['airline_sentiment']

#
### text pre-processing
#
def cleanText(text):
    
    text = re.sub(r'@[A-Za-z0-9]+', '',text) # remove @mentions
    text = re.sub(r'#', '', text) # remove the # symbol
    text = re.sub(r'RT[\s]+', '', text) # remove retweets
    text = re.sub(r'https?:\/\/\S+', '', text) # remove the hyperlink
    text = text.translate(str.maketrans('', '', string.punctuation)) # remove punctuation
    text = text.lower() # convert to lowercase
    text_tokens = word_tokenize(text)
    text = [word for word in text_tokens if not word in stop] # remove stop words
    stemmed = []
    for w in text:
        stemmed.append(ps.stem(w))

    def listToString(s):   
        str1 = " " 
        return (str1.join(s))     
        
    text = listToString(stemmed)
    return text 

#
### feature weighting
#

### tf-idf bag of words
def bagOfWordsTrain(text): # vectorise training text
    return vectorizer.fit_transform(text)

def bagOfWordsPredict(text): # vectorise prediction text
    return vectorizer.transform(text)

### binary representation
def binaryTrain(text): # vectorise training data
    return cv.fit_transform(text) 

def binaryPredict(text): # vectorise prediction data
    return cv.transform(text)

### count based representation
def countTrain(text): # vectorise training data
    return cb.fit_transform(text) 

def countPredict(text): # vectorise prediction data
    return cb.transform(text)

#
### word2vec
#
def getTokens(text): # tokenize
        tokens = []
        for tweet in text:
            tokens.append(word_tokenize(tweet))
        return tokens

def createModel(tokens): # create model
    return (Word2Vec(tokens, size = modelSize, min_count=1))
  
def buildWordVector(text, size, model): # build vectors function
    vec = np.zeros(size).reshape((1, size))
    count = 0.
    for word in text:
        try:
            vec += model[word].reshape((1, size))
            count += 1.
        except KeyError:
            continue
    if count != 0:
        vec /= count
    return vec 

def getVectors(text, size, model): # get vectors
    return np.concatenate([buildWordVector(z, size, model) for z in text])

#
### classifiers 
# 
### logistic regression
def logReg(train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec):
    clf_log = LogisticRegression()
    clf_log.fit(train_x_vectors, train_y) # train classifier
    return(predict(clf_log, train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec)) # return prediction results

### linear svm
def linSvm(train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec): 
    clf_svm = svm.SVC(kernel='linear', probability=True)
    clf_svm.fit(train_x_vectors, train_y) # train classifier
    return(predict(clf_svm, train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec)) # return prediction results

### decision tree
def decTree(train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec):
    clf_dec = DecisionTreeClassifier()
    clf_dec.fit(train_x_vectors, train_y)  # train classifier
    return(predict(clf_dec, train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec)) # return prediction results

### multinomial naive bayes
def naiveBayes(train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec):
    clf_gnb = MultinomialNB()
    clf_gnb.fit(train_x_vectors, train_y)  # train classifier
    return(predict(clf_gnb, train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec)) # return prediction results

### multi-layer perceptron
def multiLayerPer(train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec):
    clf_mlp = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
    clf_mlp.fit(train_x_vectors, train_y)  # train classifier
    return(predict(clf_mlp, train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec)) # return prediction results

### stochastic gradient descent
def stochasticGradDes(train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec):
    clf_sgd = SGDClassifier(loss="hinge", penalty="l2", max_iter=5)
    clf_sgd.fit(train_x_vectors, train_y) # train classifier
    return(predict(clf_sgd, train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec)) # return prediction results

### combined voting classifier
def combinedVote(train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec, votingType):
    
    # classifer 1: logistic regression
    clf_log = LogisticRegression()
    clf_log.fit(train_x_vectors, train_y) # train classifier 
    
    # classifier 2: linear svm
    clf_svm = svm.SVC(kernel='linear',probability=True)
    clf_svm.fit(train_x_vectors, train_y) # train classifier
    
    # combined classifier
    if (votingType == 0): # soft
        clf_comb = VotingClassifier(estimators=[('logReg', clf_log), ('linSvm', clf_svm)], voting ='soft')
        clf_comb.fit(train_x_vectors, train_y) # train classifier
        return(predict(clf_comb, train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec)) # return prediction results
    elif (votingType ==1): # hard
        clf_comb = VotingClassifier(estimators=[('logReg', clf_log), ('linSvm', clf_svm)], voting ='hard')
        clf_comb.fit(train_x_vectors, train_y) # train classifier
        return(predict(clf_comb, train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec)) # return prediction results
    
### prediction (for use in classifier)
def predict(classifier, train_x_vectors, train_y, predict_x, predict_x_vectors, predict_y, isWord2Vec):
    
    if (isWord2Vec == 0):
    
        print("Tweet 1:", predict_x[1]) # print tweet
        print("Sentiment:", classifier.predict(predict_x_vectors[1]),"\n") # predict sentiment of tweet

        print("Tweet 2:", predict_x[2]) # print tweet
        print("Sentiment:", classifier.predict(predict_x_vectors[2]),"\n") # predict sentiment of tweet

    truth = pd.DataFrame(predict_y) # df for true sentiment
    prediction_log = pd.DataFrame({'airline_sentiment':classifier.predict(predict_x_vectors)}) # df for predicted sentiment
    results_log = pd.concat([truth,prediction_log], axis =1) # dataframe for results
    results_log.columns = ['Expected','Predicted'] # rename columns
    
    accuracy = round(accuracy_score(results_log['Expected'],results_log['Predicted']),2)
    precision = round(precision_score(results_log['Expected'],results_log['Predicted'], average ='weighted'),2)
    recall = round(recall_score(results_log['Expected'],results_log['Predicted'], average ='weighted'),2)
    f1Score = round(f1_score(results_log['Expected'],results_log['Predicted'], average ='weighted'),2)
    
    cf_matrix = plot_confusion_matrix(classifier, predict_x_vectors, predict_y, cmap=plt.cm.Reds)
    plt.show() # confusion matrix plot
    print(classification_report(results_log['Expected'],results_log['Predicted'])) # classification report
    
    return [accuracy, precision, recall, f1Score] # results to be put into graph
    
#
### results graph
#
def resultsGraph(results, labels, title, numMethods):
    
    if (numMethods == 0): # if just for 1 method

        ### data
        accuracy = [results[0], results[1]]
        precision = [results[2], results[3]]
        recall = [results[4], results[5]]
        f1S = [results[6], results[7]]

        ### labels and bars
        x = np.arange(len(labels)) # label locations
        width = 0.2 # width of the bars

        fig, ax = plt.subplots()
        rects1 = ax.bar(x - 1.5*width, accuracy, width, label='Accuracy')
        rects2 = ax.bar(x - 0.5*width, precision, width, label='Precision')
        rects3 = ax.bar(x + 0.5*width, recall, width, label='Recall')
        rects4 = ax.bar(x + 1.5*width, f1S, width, label='F1 Score')

        ### title and axis labels
        ax.set_ylabel('Score')
        ax.set_title(title)
        ax.set_xticks(x)
        ax.set_xticklabels(labels)
        ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
        fancybox=True, shadow=True, ncol=5)


        def autolabel(rects):
            ### Attach a text label above each bar in *rects*, displaying its height
            for rect in rects:
                height = rect.get_height()
                ax.annotate('{}'.format(height),
                            xy=(rect.get_x() + rect.get_width() / 2, height),
                            xytext=(0, 3),  # 3 points vertical offset
                            textcoords="offset points",
                            ha='center', va='bottom')
        autolabel(rects1)
        autolabel(rects2)
        autolabel(rects3)
        autolabel(rects4)
        
        plt.show()
        
```

## Final prediction: tf-idf bag of words, logReg + svm (soft, without preprocessing)

```python
#
### combined method: bag of words representation, logReg + linSvm (soft)
#
def finalPrediction():
    
    # get training tweets and sentiments
    train_x = getText(train)
    train_y = getSentiment(train)

    # get prediction tweets and sentiments
    predict_x = getText(test)
    predict_y = getSentiment(test)

    # clean tweets
    train_x = train_x.apply(cleanText)
    predict_x = predict_x.apply(cleanText)
    
    # vectorise tweets
    train_x_vectors = bagOfWordsTrain(train_x)
    predict_x_vectors = bagOfWordsPredict(predict_x) 

    # classifier 1: logistic regression
    clf_log = LogisticRegression()
    clf_log.fit(train_x_vectors, train_y) # train classifier
    
    # classifier 2: linear svm
    clf_svm = svm.SVC(kernel='linear',probability=True)
    clf_svm.fit(train_x_vectors, train_y) # train classifier
    
    # combined voting classifier
    clf_comb = VotingClassifier(estimators=[('logReg', clf_log), ('linSvm', clf_svm)], voting ='soft')
    clf_comb.fit(train_x_vectors, train_y) # train classifier
    
    # predict tweet sentiments and create dataframe
    tweets = pd.DataFrame(getText(test)) # get tweets
    prediction_log = pd.DataFrame({'airline_sentiment':clf_comb.predict(predict_x_vectors)}) # df for predicted sentiment
    results_log = pd.concat([tweets,prediction_log], axis =1) # dataframe for results
    results_log.columns = ['Tweet','Sentiment'] # rename columns
    
    return results_log
```
